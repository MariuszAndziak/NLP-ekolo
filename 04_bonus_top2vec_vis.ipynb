```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from top2vec import Top2Vec
import top2vec
from sklearn.metrics import silhouette_score
```

```python
from babyplots import Babyplot
import umap
```

```python
df_all = pd.read_hdf('outputs/budzety_clean_topics_top2vec.h5', key='df_all')
```

```python
# Ładowanie wcześniej wytrenowanego modelu
model = Top2Vec.load('models/main_top2vec_topic.model')
```

```python
N_TOPICS = 20
model.hierarchical_topic_reduction(N_TOPICS)
```

```python
# wyciągamy reprezentację wektorową z dokumentów
X = model._get_document_vectors()
y = df_all['topics'].values
```

```python
X.shape
```

```python
# redukujemy do 3 wymiarów
reducer = umap.UMAP(random_state=0, n_components=3)
reducer.fit(X)
```

```python
emb = reducer.transform(X)
```

```python
emb
```

```python
bp = Babyplot()
bp.add_plot(emb.tolist(), 'pointCloud',
           'categories', y.tolist(), {
               'colorScale': 'Set2',
               'showLegend': True,
               'folded': True,
               'showAxes': [True, True, True],
               'axisLabels': ['x', 'y', 'z']
           })
```

```python
bp
```

```python
# Wartość metryki (w granicach 0) wskazuje na to, że klastry na siebie zachodzą
silhouette_score(X, y)
```
