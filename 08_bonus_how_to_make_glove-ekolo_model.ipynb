```python
def from_dict_to_word2vec(output_file, vector_size, vocab, binary=True):
    total_vec = len(vocab)
    
    mode = 'wb' if binary else 'w'
    with open(output_file, mode) as f:
        print(total_vec, vector_size)
        f.write(to_utf8("{} {}\n".format(total_vec, vector_size)))
        
        for word, row in tqdm(vocab.items()):
            if binary:
                row = np.array(row, dtype=np.float32)
                f.write(to_utf8(word) + b" " + row.tostring())
            else:
                f.write(to_utf8("{} {}\n".format(word, ' '.join(repr(val) for val in row))))
```

```python
def save_new_model_for_given_tokens(model, ts, output_file):
    valid_tokens = [t for t in ts if t in model]
    # print(valid_tokens)
    vectors = model[valid_tokens]
    
    ratio = vectors.shape[0] / len(ts) * 100
    print("valid vectors: {}, all: {}, ratio: {}".format(vectors.shape[0], len(ts),  ratio))

    size_vector = vectors[0].shape[0]
    
    vocab = {key: vector for key, vector in zip(valid_tokens, vectors)}
    from_dict_to_word2vec(output_file=output_file, vector_size=size_vector, vocab=vocab)
```

```python
# downloaded from https://github.com/sdadas/polish-nlp-resources#glove
model = KeyedVectors.load_word2vec_format("models/glove_100_3_polish.txt")
```

```python
save_new_model_for_given_tokens(model, slownik, 'models/glove_ekolo.bin')
```
